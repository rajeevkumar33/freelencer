```python
from sklearn.preprocessing import LabelEncoder
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

```python
export_new.drop(['Sales of Wheat in tons'],inplace=True,axis=1)
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(export_new)
X = pca.transform(export_new)
```

```python
pca.explained_variance_ratio_
```

```python
export_new_pca = pd.DataFrame(X ,columns=['pca1','pca2'])
```

```python
export_new_pca
```

```python
plt.figure(figsize=[14,14])
plt.scatter(export_new_pca['pca1'], export_new_pca['pca2'],c='green', alpha=0.5,marker='*')
plt.show()
```

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(export_new_pca)
    export_new_pca["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
connectivity = KMeans(n_clusters=3, random_state=140)
y_predict = connectivity.fit_predict(export_new_pca)
```

```python
y_predict
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=3)
# Fitting with inputs
kmeans = kmeans.fit(export_new_pca)
# Predicting the clusters
labels = kmeans.predict(export_new_pca)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
plt.figure(figsize=[14,14])
plt.scatter(export_new_pca['pca1'], export_new_pca['pca2'],c=labels, alpha=0.5,marker='*',s=100)
plt.scatter(C[:,0], C[:,1],c='red', alpha=0.5,marker='*',s=1000)
plt.show()
```

```python
import time as time
from sklearn.cluster import AgglomerativeClustering
# Compute clustering
print("Compute unstructured hierarchical clustering...")
st = time.time()
ward = AgglomerativeClustering(n_clusters=3, linkage='ward').fit(export_new_pca)
elapsed_time = time.time() - st
label = ward.labels_
print("Elapsed time: %.2fs" % elapsed_time)
print("Number of points: %i" % label.size)
```

```python
plt.figure(figsize=[14,14])
plt.scatter(export_new_pca['pca1'], export_new_pca['pca2'],c=ward.labels_, alpha=0.5,marker='*',s=100)
plt.show()
```

```python
from sklearn.neighbors import kneighbors_graph
connectivity = kneighbors_graph(export_new_pca, n_neighbors=10, include_self=False)
```

```python
print("Compute structured hierarchical clustering...")
st = time.time()
ward = AgglomerativeClustering(n_clusters=6, connectivity=connectivity,linkage='ward').fit(export_new_pca)
elapsed_time = time.time() - st
label = ward.labels_
print("Elapsed time: %.2fs" % elapsed_time)
print("Number of points: %i" % label.size)
```

```python
children = ward.children_
children
```

```python
plt.figure(figsize=[14,14])
plt.scatter(export_new_pca['pca1'], export_new_pca['pca2'],c=ward.labels_, alpha=0.5,marker='*',s=100)
plt.scatter(children[:,0], children[:,1],c='red', alpha=0.5,marker='*',s=500)
plt.show()
```
